# Carga automatizada de archivos CSV

Para evitar la escritura manual de rutas y reducir errores, se utiliza `pathlib` para identificar automáticamente todos los archivos `.csv` disponibles en la carpeta `data`. Esta decisión mejora la reproducibilidad del notebook, ya que permite incorporar nuevos archivos anuales sin modificar manualmente el código de carga.

Además, se define una función para cargar y consolidar los archivos en un único DataFrame. Esto permite aplicar un flujo uniforme sobre los datos del período 2020-2025, manteniendo trazabilidad mediante la columna `archivo_origen`, que identifica desde qué archivo proviene cada registro.

```python
from pathlib import Path
import pandas as pd
import numpy as np

# Ruta base donde se encuentran los archivos CSV.
DATA_DIR = Path("../data")

# Busca automáticamente todos los archivos CSV dentro de la carpeta data.
archivos_csv = sorted(DATA_DIR.glob("*.csv"))

print(f"Archivos CSV encontrados: {len(archivos_csv)}")
```

```python

def cargar_archivos_csv(rutas):
    """
    Carga múltiples archivos CSV y los combina en un solo DataFrame.

    Parámetros:
    rutas: list
        Lista de rutas de archivos CSV.

    Retorna:
    pandas.DataFrame
        DataFrame consolidado con todos los archivos cargados.
    """
    dataframes = []

    for ruta in rutas:
        df_temp = pd.read_csv(ruta)
        df_temp["archivo_origen"] = ruta.name
        dataframes.append(df_temp)

    df_consolidado = pd.concat(dataframes, ignore_index=True)

    return df_consolidado

def resumen_dataset(df):
    """
    Genera un resumen inicial del conjunto de datos.
    """
    resumen = {
        "filas": df.shape[0],
        "columnas": df.shape[1],
        "duplicados": df.duplicated().sum(),
        "columnas_disponibles": list(df.columns)
    }

    return resumen

def revisar_valores_nulos(df):
    """
    Calcula la cantidad y porcentaje de valores nulos por columna.
    """
    valores_nulos = df.isnull().sum()
    porcentaje_nulos = (valores_nulos / len(df)) * 100

    resumen_nulos = pd.DataFrame({
        "valores_nulos": valores_nulos,
        "porcentaje_nulos": porcentaje_nulos.round(2)
    })

    return resumen_nulos.sort_values(by="porcentaje_nulos", ascending=False)

def revisar_tipos_datos(df):
    """
    Retorna los tipos de datos de cada columna del DataFrame.
    """
    return pd.DataFrame({
        "columna": df.columns,
        "tipo_dato": df.dtypes.astype(str).values
    })
```

```python
df = cargar_archivos_csv(archivos_csv)
df.head()
```

```python
resumen_dataset(df)
```

```python
revisar_valores_nulos(df)
```

## Justificación técnica del uso de funciones

El uso de funciones permite estructurar el flujo inicial de trabajo de manera más clara y reproducible. En esta fase, las funciones cumplen tres propósitos principales: cargar automáticamente los archivos CSV, consolidar los datos en un único DataFrame y generar resúmenes técnicos sobre la estructura y calidad inicial del dataset.

Esta estrategia responde al principio de coherencia entre entrada, procesamiento y resultado. La entrada corresponde a los archivos CSV anuales del WFP; el procesamiento se realiza mediante funciones de carga, consolidación y revisión; y los resultados corresponden a un DataFrame unificado, dimensiones del dataset, cantidad de duplicados, columnas disponibles, tipos de datos y valores nulos.

Además, esta estructura facilita que cualquier integrante del equipo pueda ejecutar nuevamente el notebook dentro del entorno configurado, reduciendo errores manuales y fortaleciendo la reproducibilidad del proyecto.
